```python
import matplotlib.pyplot as plt

# 1. Major Category Distribution
plt.figure(figsize=(8,5))
df["Major_Category"].value_counts().plot(kind="bar")
plt.title("Students by Major Category")
plt.xlabel("Major Category")
plt.ylabel("Number of Students")
plt.show()


```

```python
## 2. Year of Study Distribution
plt.figure(figsize=(8,5))
df["Year_of_Study"].value_counts().plot(kind="bar")
plt.title("Students by Year of Study")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# 5. GPA Comparison
avg_pre = df["Pre_Semester_GPA"].mean()
avg_post = df["Post_Semester_GPA"].mean()

plt.figure(figsize=(5,4))
plt.bar(["Pre GPA", "Post GPA"], [avg_pre, avg_post])
plt.title("Average GPA Comparison")
plt.ylabel("GPA")
plt.show()
```

```python
# 4. Burnout Risk Level
plt.figure(figsize=(6,4))
df["Burnout_Risk_Level"].value_counts().plot(kind="bar")
plt.title("Burnout Risk Level")
plt.xlabel("Risk Level")
plt.ylabel("Number of Students")
plt.show()
```

```python
# 5. GPA Comparison
avg_pre = df["Pre_Semester_GPA"].mean()
avg_post = df["Post_Semester_GPA"].mean()

plt.figure(figsize=(5,4))
plt.bar(["Pre GPA", "Post GPA"], [avg_pre, avg_post])
plt.title("Average GPA Comparison")
plt.ylabel("GPA")
plt.show()
```
